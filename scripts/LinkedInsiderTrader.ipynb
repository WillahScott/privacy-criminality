# SEC Insider Trader and LinkedIn

```python
import pandas as pd
import re
```

```python
# Read data
SEC_data = pd.read_csv('data/SEC_filings.csv')
SEC_data.rename(columns = {'Unnamed: 0':'rowID', 'Insider Relationship':'InsiderRelationship'}, inplace=True)

# Get only Row, Company and Insider Trader
traders = SEC_data[['rowID','Company','InsiderRelationship']]
```

```python
traders
```

Let's get some names...

```python
# Remove certain keywords
remove = "10%|&|CEO|CFO|CIO|COO|President|Director|Owner|Chairman|Chief|Officer|Executive|Operating|and|Vice|VP|EVP|VICE|PRESIDENT|Counsel|Secretary|General|Other"

# Parse name (split into columns, return first or last name)
def get_name(text, index):
    words = re.sub(remove, '', text).split()
    return words[index]

# Create a first name and a last name column
traders.loc[:,'FirstName'] = pd.Series(traders['InsiderRelationship'].apply( lambda x: get_name(x,1) ))
traders.loc[:,'LastName'] = pd.Series(traders['InsiderRelationship'].apply( lambda x: get_name(x,0) ))
```
